```python
#Transform

import pandas as pd

#Gathering CSV data
orders_data=pd.read_csv('orders.csv')
#orders_data.head(20)

#What I observed?
# 1) Observed Ship Mode column consists of 3 Types of null values ( Not Available, Null,unknown)
# 2) Convert Column names to lower case and need to replace the spaces with underscore
orders_data['Ship Mode'].unique() #array(['Second Class', 'Standard Class', 'Not Available', 'unknown','First Class', nan, 'Same Day'], dtype=object)
#Observed Ship Mode column consists of 3 Types of null values ( Not Available, Null,unknown)
#Planning to remove them while csv data loading
```

```python
orders_data=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
orders_data['Ship Mode'].unique()
#Output : array(['Second Class', 'Standard Class', nan, 'First Class', 'Same Day'],dtype=object)
```

```python
# 2) Convert Column names to lower case and need to replace the spaces with underscore
orders_data.columns

'''Output : Index(['Order Id', 'Order Date', 'Ship Mode', 'Segment', 'Country', 'City',
       'State', 'Postal Code', 'Region', 'Category', 'Sub Category',
       'Product Id', 'cost price', 'List Price', 'Quantity',
       'Discount Percent'],
      dtype='object')'''
```

```python
orders_data.columns=orders_data.columns.str.lower()
```

```python
orders_data.columns=orders_data.columns.str.replace(' ','_')
```

```python
orders_data
```

```python
#datatype conversion
#orders_data.dtypes
#Need to convert order_date from object to Date
orders_data['order_date']=pd.to_datetime(orders_data['order_date'],format="%Y-%m-%d")
```

```python
#Here we have Discount% but there is not sale_price,need to find that
orders_data['discount']=orders_data['list_price']*orders_data['discount_percent']*0.01
orders_data['sale_price']=orders_data['list_price']-orders_data['discount']

#Profit
orders_data['profit']=orders_data['sale_price']-orders_data['cost_price']
```

```python
orders_data
```

```python
#dropping the columns which are not required
orders_data.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
